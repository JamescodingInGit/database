Feature Selection Part

```python
products.columns
```

```python
# products = products[['product_id', 'product_name', 'category', 'about_product','product_link', 'img_link', 'rating']]
products = products[['product_id', 'product_name', 'category', 'about_product','product_link', 'img_link', 'rating', 'actual_price']]
```

```python
products
```

```python
products['tags'] = products['category']+products['about_product']
```

```python
products
```

```python
# new_data = products.drop(columns=['category','about_product'])
new_data = products
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=1465, stop_words='english')
```

```python
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
# new_data[new_data['product_name']=="Wayona Nylon Braided USB to Lightning Fast Charging and Data Sync Cable Compatible for iPhone 13, 12,11, X, 8, 7, 6, 5, iPad Air, Pro, Mini (3 FT Pack of 1, Grey)"]
new_data[new_data['product_id']=="B096MSW6CT"].index[0]
```

```python
new_data.info()
```

```python
distance = sorted(list(enumerate(similarity[0])),reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
     print(new_data.iloc[i[0]].product_name)
```

```python
# test
distance = sorted(list(enumerate(similarity[0])), reverse=True, key=lambda vector: vector[1])
recommended_products = set()

for i in distance:
    product_index = i[0]
    product_name = new_data.iloc[product_index].product_name
    if product_name not in recommended_products:
        print(product_name)
        recommended_products.add(product_name)
        if len(recommended_products) == 5:  # Stop when 5 unique recommendations are found
            break
```

```python
# recommand("Wayona Nylon Braided USB")
```

```python
def recommend(products):
    matches = new_data[new_data['product_name'].str.contains(products, case=False)]
    if not matches.empty:
        index = matches.index[0]
        distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector: vector[1])
        for i in distance[0:5]:
            print(new_data.iloc[i[0]].product_name)
    else:
        print("Product not found in the dataset")
```

```python
recommend("pTron Solero")
```

```python
# Convert 'rating' column to numeric type, ignore errors
products['rating'] = pd.to_numeric(products['rating'], errors='coerce')
```

```python
# Calculate the mean rating
calculate_rating = products['rating'].mean()
print(calculate_rating)
```

```python
mean_ratings = products.groupby('product_name')['rating'].mean()
sorted_brands = mean_ratings.sort_values(ascending=False)
print("Brands sorted by mean rating:")
print(sorted_brands)
```
